Array

```python
import numpy as np

Value=[2, 4, 8, 10, 18, 20, 45, 64, 76, 81, 96, 100]

result = np.array(Value)

print(f"NumPy Array: {result}")
print(f"Type: {type(result)}")
print(f"Data Type: {result.dtype}")
```

user Input

```python
import numpy as np
values = list(map(int, input("Enter numbers separated by space: ").split()))

result = np.array(values)

print(f"NumPy Array: {result}")
print(f"Type: {type(result)}")
```

1-D Arrays

```python
import numpy as np
Value=[2, 4, 8, 10, 18, 20, 45, 64, 76, 81, 96, 100]

result = np.array(Value)

print("One-Dimensional Array")
print(f"Result: {result}")
print(f"Array Dimensional: {result.ndim}D")
print(f"Array Shape: {result.shape}")
```

2-D Arrays

```python
import numpy as np
Value=[[2, 4, 8, 10, 18, 20], [45, 64, 76, 81, 96, 100]]

result = np.array(Value)

print("Two-Dimensional Array")
print(f"Result: \n{result}")
print(f"Array Dimensional: {result.ndim}D")
print(f"Array Shape: {result.shape}")
```

3-D Arrays

```python
import numpy as np
Value=[[2, 4, 8, 10], [18, 20, 45, 64], [76, 81, 96, 100]]

result = np.array(Value)

print("Three-Dimensional Array")
print(f"Result: \n{result}")
print(f"Array Dimensional: {result.ndim}D")
print(f"Array Shape: {result.shape}")
```

Multi-D Array

```python
import numpy as np
Value=[2, 4, 8, 10, 18, 20, 45, 64, 76, 81, 96, 100]

result = np.array(Value, ndmin=6)

print("Multi-Dimensional Array")
print(f"Result: \n{result}")
print(f"Array Dimensional: {result.ndim}D")
```
