Zeros

```python
import numpy as np
Value = (4, 6)

result = np.zeros(Value, dtype=int)

print(f"Zeros Array: \n{result}")
print(f"Array Shape: {result.shape}")
```

Ones

```python
import numpy as np
Value = (4, 6)

result = np.ones(Value, dtype=int)

print(f"Ones Array: \n{result}")
print(f"Array Shape: {result.shape}")
```

Empty

```python
import numpy as np
Value = (4, 6)

result = np.empty(Value, dtype=int)

print(f"Empty Array: \n{result}")
print(f"Array Shape: {result.shape}")
```

Range

```python
import numpy as np
Value = (10+1)

result = np.arange(Value, dtype=int)

print(f"Range Array: \n{result}")
print(f"Array Shape: {result.shape}")
```

Diagonal

```python
import numpy as np
Value = (4, 6)

result = np.eye(*Value, dtype=int)

print(f"Diagonal Array: \n{result}")
print(f"Array Shape: {result.shape}")
```

LinSpace

```python
import numpy as np
result = np.linspace(4, 40, num=10)

print(f"LinSpace Array: \n{result}")
print(f"Array Shape: {result.shape}")
```
