Python Lists

```python
# Measure execution time of raising numbers 1–10 to the 4th power
%timeit [j**4 for j in range(1, 11)]

# Store and display the result
result = [j**4 for j in range(1, 11)]
print(f"Result: {result}")
```

NumPy Array

```python
import numpy as np

# Measure execution time using NumPy
%timeit np.arange(1, 11) ** 4

# Store and print the result
result = np.arange(1, 11) ** 4
print(f"Result: {result}")
```
