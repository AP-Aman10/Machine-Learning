Histogram Plot

```python
import matplotlib.pyplot as plt

values = [88, 63, 24, 58, 74, 41, 39, 39, 61, 25, 11, 75, 73, 96, 37, 74, 36, 70, 13, 26,
          40, 17, 98, 11, 41, 76, 85, 11, 47, 24, 93, 38, 12, 59, 28, 100, 20, 44, 60, 95,
          29, 36, 94, 80, 25, 87, 16, 86, 65, 63]

Col = ["blue", "red", "green", "yellow", "orange", "purple",
       "pink", "brown", "gray", "cyan", "magenta", "turquoise"]

# Explicit bins (no gaps)

plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(values, edgecolor="black", linestyle="-.")

for patch, color in zip(patches, Col):
    patch.set_facecolor(color)

plt.title("Distribution of Values", fontsize=16)
plt.xlabel("Value Range", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(range(10, 100, 10))

plt.tight_layout()
plt.grid(False)
plt.show()
```
