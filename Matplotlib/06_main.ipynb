Pie Plot

```python
import matplotlib.pyplot as plt

languages = ["HTML", "JS", "C++", "Python", "C", "CSS", "Java"]
popularity = [20, 81, 10, 100, 64, 45, 96]

col = ["yellow", "Blue", "Green", "Red", "Orange", "Purple", "Pink"]
plt.pie(
        popularity,
        labels=languages,
        colors=col,
        autopct="%0.1f%%",
        shadow=True, radius=2,
        counterclock=False,
        textprops={"fontsize":15},
        wedgeprops={"linewidth":2, "edgecolor":"Black"})
plt.legend(loc=4)
plt.show()
```

```python
import matplotlib.pyplot as plt

languages = ["HTML", "JS", "C++", "Python", "C", "CSS", "Java"]
popularity = [20, 81, 10, 100, 64, 45, 96]

colors = ["yellow", "blue", "green", "red", "orange", "purple", "pink"]

plt.figure(figsize=(8, 8))

# Pie chart
plt.pie(
    popularity,
    labels=languages,
    colors=colors,
    autopct="%.1f%%",
    startangle=140,
    pctdistance=0.8,
    textprops={"fontsize": 12},
    shadow=True,
    wedgeprops={"edgecolor": "black", "linewidth": 2}
)

# White center (donut hole)
centre_circle = plt.Circle((0, 0), 0.60, fc="white")
plt.gca().add_artist(centre_circle)

plt.title("Programming Language Popularity", fontsize=16, fontweight="bold")
plt.legend()
plt.show()
```
