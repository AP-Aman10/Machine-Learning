Subplot Plot

```python
import matplotlib.pyplot as plt

# -------------------- Data --------------------
x_values = list(range(1, 11))
y_values = [4, 96, 10, 64, 45, 20, 8, 81, 100, 2]

languages = ["HTML", "JS", "C++", "Python", "C", "CSS", "Java"]
popularity = [20, 81, 10, 100, 64, 45, 96]
colors = ["yellow", "blue", "green", "red", "orange", "purple", "pink"]
markers = ["h", "p", "d", "H", "D", "*", "s"]

# -------------------- Figure Setup --------------------
plt.figure(figsize=(12, 10))

# -------------------- Subplot 1: Step Plot --------------------
plt.subplot(2, 2, 1)
plt.step(
    x_values,
    y_values,
    color="black",
    marker="h",
    markerfacecolor="red",
    label="Data Values"
)
plt.title("Step Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# -------------------- Subplot 2: Scatter Plot --------------------
plt.subplot(2, 2, 2)
for lang, pop, col, mark in zip(languages, popularity, colors, markers):
    plt.scatter(
        lang,
        pop,
        color=col,
        marker=mark,
        s=120,
        edgecolors="black"
    )

plt.title("Programming Language Popularity (Scatter)")
plt.xlabel("Languages")
plt.ylabel("Popularity")

# -------------------- Subplot 3: Pie Chart --------------------
plt.subplot(2, 2, 3)
plt.pie(
    popularity,
    labels=languages,
    colors=colors,
    autopct="%0.1f%%",
    shadow=True,
    startangle=90,
    textprops={"fontsize": 12},
    wedgeprops={"linewidth": 2, "edgecolor": "black"}
)
plt.title("Popularity Distribution (Pie Chart)")

# -------------------- Subplot 4: Bar Chart --------------------
plt.subplot(2, 2, 4)
plt.bar(
    languages,
    popularity,
    color=colors,
    edgecolor="black",
    linestyle="-."
)
plt.title("Programming Language Popularity (Bar Chart)")
plt.xlabel("Languages")
plt.ylabel("Popularity")

# -------------------- Final Layout --------------------
plt.suptitle("Matplotlib Subplots Example", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```
