Scatter Plot

```python
import matplotlib.pyplot as plt

# Data
languages = ["HTML", "JS", "C++", "Python", "C", "CSS", "Java"]
popularity = [20, 81, 10, 100, 64, 45, 96]

colors = ["yellow", "blue", "green", "red", "orange", "purple", "pink"]
markers = ["h", "p", "d", "H", "D", "*", "s"]

# Create figure
plt.figure(figsize=(10, 6))

# Plot points
for lang, pop, color, marker in zip(languages, popularity, colors, markers):
    plt.scatter(
        lang,
        pop,
        color=color,
        marker=marker,
        s=120,
        label=lang,
        edgecolors="black"
    )

# Labels and title
plt.title("Programming Language Popularity (2026)", fontsize=14, fontweight="bold")
plt.xlabel("Programming Languages", fontsize=12)
plt.ylabel("Popularity Score", fontsize=12)

# Extras
plt.legend(title="Languages")
plt.ylim(0, 110)

plt.show()
-
```
