Box & Whisker plot

```python
import matplotlib.pyplot as plt

value = [4, 8, 10, 20, 45, 64, 81, 96, 100]

plt.figure(figsize=(6, 4))

plt.boxplot(
    value,
    notch=True,
    patch_artist=True,
    labels=["Values"],
    boxprops=dict(facecolor="skyblue", color="black"),
    medianprops=dict(color="red", linewidth=1.4),
    whiskerprops=dict(color="black"),
    showmeans= True,
    capprops=dict(color="black"),
    meanprops=dict(marker="h", markerfacecolor="green", markeredgecolor="black")
)

plt.title("Box Plot of Values", fontsize=14)
plt.ylabel("Value", fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt

Value1 = [4, 8, 10, 20, 45, 64, 81, 96, 100]
Value2 = [2, 12, 18, 22, 47, 69, 84, 94, 100]

data = [Value1, Value2]

plt.figure(figsize=(6, 4))

box = plt.boxplot(
    data,
    notch=True,
    patch_artist=True,
    labels=["Value Set 1", "Value Set 2"],
    showmeans=True,
    boxprops=dict(color="black"),
    medianprops=dict(color="red", linewidth=1.4),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
    meanprops=dict(marker="h", markerfacecolor="green", markeredgecolor="black")
)

# Set different colors for each box
colors = ["skyblue", "lightcoral"]
for patch, color in zip(box["boxes"], colors):
    patch.set_facecolor(color)

plt.title("Box Plot of Values", fontsize=14)
plt.ylabel("Value", fontsize=12)

plt.show()
```
