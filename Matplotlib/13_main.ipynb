```python
import matplotlib.pyplot as plt
import numpy as np

languages = ["HTML", "JS", "C++", "Python", "C", "CSS", "Java"]
popularity = [20, 81, 10, 100, 64, 45, 96]
Using = [10, 20, 45, 64, 81, 96, 100]

width = 0.35
x = np.arange(len(languages))

plt.figure(figsize=(10, 6))

plt.bar(x - width/2, popularity, width,
        label="Popularity",
        color="red",
        edgecolor="black",
        linestyle="-.")

plt.bar(x + width/2, Using, width,
        label="Using",
        color="blue",
        edgecolor="black",
        linestyle="-.")

plt.title("Programming Language Popularity (2026)", fontsize=18, fontweight="bold")
plt.xlabel("Programming Languages", fontsize=12)
plt.ylabel("Popularity Score", fontsize=12)

plt.xticks(x, languages)

plt.savefig("SaveFig")
plt.legend()
plt.grid(False)
plt.show()
```
